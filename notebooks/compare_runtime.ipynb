```python
import matplotlib.pyplot as plt
import matplotlib as mpl
elapsed_time = [94, 41+4*60, 43*60, (20*60+28)*60, 280, 89]
methods = ['T1SEstacker', 'T3SEpp', 'CNN-T4SE', 'Bastion6', 'EffectiveDB', 'DeepSecE']

mpl.rcParams['font.family'] = 'Arial'
mpl.rcParams['axes.labelsize'] = 12
mpl.rcParams['axes.titlesize'] = 14
mpl.rcParams['legend.title_fontsize'] = 11
mpl.rcParams['legend.fontsize'] = 11
mpl.rcParams['xtick.labelsize'] = 11
mpl.rcParams['ytick.labelsize'] = 11
```

```python
import numpy as np
import seaborn as sns

plt.bar(methods, elapsed_time, width=0.5)
plt.yscale('log')
plt.ylim([30, 24*60*60])
plt.yticks([10, 100, 1000, 10000, 100000])

horizontal_lines = [60, 600, 3600]  # Example line positions
for line in horizontal_lines:
    plt.axhline(y=line, color='black', alpha=0.5, linestyle='--')
    
plt.ylabel('Elapsed time (s)')
plt.xlabel('Method')
plt.title('Predict 260 test sequences')


sns.despine()
plt.tight_layout()
```

```python
import numpy as np
import seaborn as sns

elapsed_time_ = [seconds/60 for seconds in elapsed_time]

plt.figure(figsize=(6, 5))
plt.bar(methods, elapsed_time_, width=0.5, color='#7dd0b6', alpha=0.7)
plt.yscale('log')
plt.ylim([0.5, 24*60])
plt.yticks([1, 10, 100, 1000], labels=['1', '10', '100', '1000'])

horizontal_lines = [1, 10, 100]  # Example line positions
for line in horizontal_lines:
    plt.axhline(y=line, color='black', alpha=0.5, linestyle='--')
    
plt.ylabel('Elapsed time (min)')
plt.xlabel('Method')
plt.title('Predict 260 test sequences')

sns.despine()
plt.tight_layout()
```
